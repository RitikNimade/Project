```python
# apply condition for filter out our finale data
Finale_rating=Rating[Rating['Rating_y']>=50]
```

```python
# find out any duplicated raw is present or not
dup = Finale_rating[Finale_rating.duplicated(['User_ID','Title'])]
```

```python
# drop out duplicates 
F_R=Finale_rating.drop_duplicates(['Title','User_ID'])
```

```python
F_R.shape
```

```python
# create a pivot table
pivot_table = F_R.pivot_table( values='Rating_y',index='Title',columns='User_ID',)
```

```python
pivot_table
```

```python
# fill nan value woth 0
p_table = pivot_table.fillna(0)
```

```python
p_table
```

```python
import scipy 
```

```python
# we can not fit out pivot table on model so we create a csr_matrix with neglet our 0 in pivot table
from scipy.sparse import csr_matrix
```

```python
book_sparse = csr_matrix(p_table)
```

```python
book_sparse
```

```python
from sklearn.neighbors import NearestNeighbors
```

```python
model = NearestNeighbors(algorithm = 'brute')
```

```python
model.fit(book_sparse)
```

```python
# take out two thingh from our model and testing model
# where we give 200 index position as input data of pivot table 
# n_neighbors gives as number of result or recommandation
Distance,Suggestion = model.kneighbors(p_table.iloc[126,:].values.reshape(1,-1), n_neighbors=6)
```

```python
# distace means wwhich book is near from all of our books
Distance
```

```python
# it gives index of recommandated book
Suggestion
```

```python
# using for loop for iteration and print the book name 
# it print the len output like suggestion and iterated them
# than the name of book is present in index of p_table so print them
# on the basis of index of suggestion
for i in range(len(Suggestion)):
    print(p_table.index[Suggestion[i]])
```

```python
# create a funtion for recommandation
def Book_Recommandation(book_i):
    book_Id = np.where(p_table.index == book_i)[0][0] # use [0][0]   for  remove list or dtype of book id
    Distance,Suggestion = model.kneighbors(p_table.iloc[book_Id,:].values.reshape(1,-1), n_neighbors=6)
    
    for i in range(len(Suggestion)):
        books = p_table.index[Suggestion[i]]
        for j in books:# this is use for print the name of book in clear form not in list 
            print(j)
```
